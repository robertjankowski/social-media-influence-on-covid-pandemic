```python
%load_ext autoreload
%autoreload 2

import sys

sys.path.append('../')

import scripts.visualization as viz
from scripts.data_utils import load_multilayer_results

viz.load_matplotlib()
plt = viz.plt
```

# Large network

## Parameters


- $\beta$ = {0.1, 0.3, 0.5, 0.7}
- $\gamma$ = 0.5
- $\mu$ = 0.85
- $\kappa$ = 0.1
- q = 3
- max_infected_time = 10
- NSTEPS = 200 000
- NAGENTS = 10 000
- NFRACLINKS = 0.1
- n_runs = 1 (only one realization !!)

```python
def get_dr_path(beta_label, beta):
    return f"../data/new_experiments/multilayer/p_beta_positive_opinion/dead_ratio_p_q_3_beta_{beta_label}_L1-beta={beta}_gamma=0.5_mu=0.85_kappa=0.1_max_infected_time=10_L2-q=4_p=0.5_xi=0.1_n=100_NRUNS=1_NSTEPS=200000_NAGENTS=10000_NFRACLINKS=0.1.csv"

dr_beta_1_params, dr_beta_1_df = load_multilayer_results(get_dr_path('0_1', 0.1))
dr_beta_3_params, dr_beta_3_df = load_multilayer_results(get_dr_path('0_3', 0.3))
dr_beta_5_params, dr_beta_5_df = load_multilayer_results(get_dr_path('0_5', 0.5))
dr_beta_7_params, dr_beta_7_df = load_multilayer_results(get_dr_path('0_7', 0.7))
dr_beta_9_params, dr_beta_9_df = load_multilayer_results(get_dr_path('0_9', 0.9))
```

```python
def get_ir_path(beta_label, beta):
    return f"../data/new_experiments/multilayer/p_beta_positive_opinion/infected_ratio_p_q_3_beta_{beta_label}_L1-beta={beta}_gamma=0.5_mu=0.85_kappa=0.1_max_infected_time=10_L2-q=4_p=0.5_xi=0.1_n=100_NRUNS=1_NSTEPS=200000_NAGENTS=10000_NFRACLINKS=0.1.csv"

ir_beta_1_params, ir_beta_1_df = load_multilayer_results(get_ir_path('0_1', 0.1))
ir_beta_3_params, ir_beta_3_df = load_multilayer_results(get_ir_path('0_3', 0.3))
ir_beta_5_params, ir_beta_5_df = load_multilayer_results(get_ir_path('0_5', 0.5))
ir_beta_7_params, ir_beta_7_df = load_multilayer_results(get_ir_path('0_7', 0.7))
ir_beta_9_params, ir_beta_9_df = load_multilayer_results(get_ir_path('0_9', 0.9))
```

```python
def get_opinion_path(beta_label, beta):
    return f"../data/new_experiments/multilayer/p_beta_positive_opinion/mean_opinion_p_q_3_beta_{beta_label}_L1-beta={beta}_gamma=0.5_mu=0.85_kappa=0.1_max_infected_time=10_L2-q=4_p=0.5_xi=0.1_n=100_NRUNS=1_NSTEPS=200000_NAGENTS=10000_NFRACLINKS=0.1.csv"

opinion_beta_1_params, opinion_beta_1_df = load_multilayer_results(get_opinion_path('0_1', 0.1))
opinion_beta_3_params, opinion_beta_3_df = load_multilayer_results(get_opinion_path('0_3', 0.3))
opinion_beta_5_params, opinion_beta_5_df = load_multilayer_results(get_opinion_path('0_5', 0.5))
opinion_beta_7_params, opinion_beta_7_df = load_multilayer_results(get_opinion_path('0_7', 0.7))
opinion_beta_9_params, opinion_beta_9_df = load_multilayer_results(get_opinion_path('0_9', 0.9))
```

```python
colors = ['red', 'green', 'blue', 'black', 'yellow', 'pink']
colors = ['xkcd:' + c for c in colors]
markers = ['o', 's', 'v', '^', 'X', 'D']
markersize = 6

labels = [fr'$\beta$={x}' for x in [0.1, 0.3, 0.5, 0.7, 0.9]]
ps = [float(x) for x in opinion_beta_1_df.columns]
ys = [opinion_beta_1_df, opinion_beta_3_df, opinion_beta_5_df, 
      opinion_beta_7_df, opinion_beta_9_df]

for i, y in enumerate(ys):
    plt.plot(ps, y.values[0], color=colors[i], label=labels[i], marker=markers[i],
             markersize=markersize)

plt.ylabel('opinion', fontsize=20)
plt.xlabel('p', fontsize=20)
plt.legend()
```

```python
ys = [dr_beta_1_df, dr_beta_3_df, dr_beta_5_df, dr_beta_7_df, dr_beta_9_df]

for i, y in enumerate(ys):
    plt.plot(ps, y.values[0], color=colors[i], label=labels[i], marker=markers[i],
             markersize=markersize)

plt.ylabel('dead rate', fontsize=20)
plt.xlabel('p', fontsize=20)
plt.legend()
```
