### $\beta$ vs $\lambda$ experiment

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scripts.visualization as viz
```

```python
# from scripts.experiments import beta_lambda_dead_experiment
# 
# betas, lambdas, dead_ratio_output, recovered_ratio_output = \
#     beta_lambda_dead_experiment(resolution=100)

def read_from_file(filename: str):
    output = []
    with open(filename, 'r') as f:
        for line in f:
            output.append(float(line))
    return output

dead_ratio_output = np.load("../data/experiments/beta_lambda_dead_ratio_output_res=100.npy")
recovered_ratio_output = np.load("../data/experiments/beta_lambda_recovered_ratio_output_res=100.npy")
lambdas = read_from_file("../data/experiments/lambda_res=100.txt")
betas = read_from_file("../data/experiments/betas_res=100.txt")
```

```python
plt.figure(figsize=(8, 6))
xticks_labels = ['{:.2f}'.format(l) for l in lambdas]
yticks_labels = ['{:.2f}'.format(b) for b in betas[::-1]]
sns.heatmap(dead_ratio_output[::-1], annot=False, cmap="YlGnBu",
            yticklabels=yticks_labels, xticklabels=xticks_labels, 
            vmin=0, vmax=1, cbar_kws={'label': 'Dead ratio'})

plt.locator_params(axis='y', nbins=10)
plt.locator_params(axis='x', nbins=10)
plt.xticks(rotation=0)
plt.ylabel(r'$\beta$', fontsize=20)
plt.xlabel(r'$\lambda$', fontsize=20)
plt.title(r'$q=4, p=0.5, \varepsilon=0.3| \delta=0.4 | \gamma=0.6, \mu=0.9$', fontsize=16)
# viz.save_figure("../plots/beta_lambda_dead_ratio_heatmap_resolution=100.pdf")
plt.show()
```
