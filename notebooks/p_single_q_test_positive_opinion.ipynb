# Large network

## Parameters


- $\beta$ = 0.7
- $\gamma$ = 0.5
- $\mu$ = 0.85
- $\kappa$ = 0.1
- max_infected_time = 10
- NSTEPS = 200 000
- NAGENTS = 10 000
- NFRACLINKS = 0.1
- n_runs = 1 (only one realization !!)

And we consider $q = \{3, 4, 5\}$

```python
def get_dr(q):
    path = f"../data/new_experiments/multilayer/p_q_positive_opinion/dead_ratio_p_{q}_q_L1-beta=0.7_gamma=0.5_mu=0.85_kappa=0.1_max_infected_time=10_L2-q=4_p=0.5_xi=0.1_n=100_NRUNS=1_NSTEPS=200000_NAGENTS=10000_NFRACLINKS=0.1.csv"
    return pd.read_csv(path, index_col=0)
    
    
dr_q3 = get_dr(3)
dr_q4 = get_dr(4)
dr_q5 = get_dr(5)
```

```python
def get_ir(q):
    path = f"../data/new_experiments/multilayer/p_q_positive_opinion/infected_ratio_p_{q}_q_L1-beta=0.7_gamma=0.5_mu=0.85_kappa=0.1_max_infected_time=10_L2-q=4_p=0.5_xi=0.1_n=100_NRUNS=1_NSTEPS=200000_NAGENTS=10000_NFRACLINKS=0.1.csv"
    return pd.read_csv(path, index_col=0)

ir_q3 = get_ir(3)
ir_q4 = get_ir(4)
ir_q5 = get_ir(5)
```

```python
def get_opinion(q):
    path = f"../data/new_experiments/multilayer/p_q_positive_opinion/mean_opinion_p_{q}_q_L1-beta=0.7_gamma=0.5_mu=0.85_kappa=0.1_max_infected_time=10_L2-q=4_p=0.5_xi=0.1_n=100_NRUNS=1_NSTEPS=200000_NAGENTS=10000_NFRACLINKS=0.1.csv"
    return pd.read_csv(path, index_col=0)

opinion_q3 = get_opinion(3)
opinion_q4 = get_opinion(4)
opinion_q5 = get_opinion(5)
```

```python
colors = ['red', 'green', 'blue', 'black']
colors = ['xkcd:' + c for c in colors]
markers = ['o', 's', 'v', 'D']
markersize = 6

labels = [f'q={x}' for x in [3, 4, 5]]
ps = [float(x) for x in opinion_q3.columns]
ys = [opinion_q3, opinion_q4, opinion_q5]

for i, y in enumerate(ys):
    plt.plot(ps, y.values[0], color=colors[i], label=labels[i], marker=markers[i],
             markersize=markersize)

plt.ylabel('opinion', fontsize=20)
plt.xlabel('p', fontsize=20)
plt.legend()
```

```python
ys = [dr_q3, dr_q4, dr_q5]

for i, y in enumerate(ys):
    plt.plot(ps, y.values[0], color=colors[i], label=labels[i], marker=markers[i],
             markersize=markersize)

plt.ylabel('dead rate', fontsize=20)
plt.xlabel('p', fontsize=20)
plt.legend()
```

```python
ys = [ir_q3, ir_q4, ir_q5]

for i, y in enumerate(ys):
    plt.plot(ps, y.values[0], color=colors[i], label=labels[i], marker=markers[i],
             markersize=markersize)

plt.ylabel('max infected rate', fontsize=20)
plt.xlabel('p', fontsize=20)
plt.legend()
```
